# Time Series

Any time series data has time component in it and x-axis in time series plots will usually denote time and y-axis the variable which is tracked over time. Eg - Rainfall data, population and stock prices.

Although time is infinitesimal but we generally measure time in terms of change at specific intervals. These could be years, months, days, hours, seconds etc. It is finite discritization of continuam.

Time series data are presented in discrete signals for analysis. The quantities can also be expressed in continuam with help of function but in that case time series modelling does not apply & it is simply regression over time. 

##  Time domain representation of time series data

When time series data is represented over time domain it is called time domain representation.

##  Frequency domain representation of time series data

The same data for a given data observation window can also be represented in frequency domain. Which in simple terms, is an analysis to capture periodicity in data.
