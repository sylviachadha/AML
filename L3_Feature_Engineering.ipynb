# <font color=blue>**1. Feature Scaling on Numerical columns**

Feature Scaling Types

- Min-max scaling
- Mean scaling
- Standardization
- Robust

You go for any feature scaling but then same feature scaling should be applied to all numerical col's in data.

The same pre-processing should be applied to both train and test data.

### a. Min-Max Scaling

You can calculate max and min values of the jth feature.

Difference in values will be on same scale for each feature so it is easy to compare.

### b. Mean Scaling

Here u r not taking difference from minimum but from mean of data.

You can calculate mean value of the jth feature.

### c. Standardization Scaling

<font color=blue> *We r taking difference of values from some standard metrics (e.g. min or mean) in NUMERATOR that is generally referred to as shifting although we r not shifting anything it just refers to how far away values are from metrics like mean or min & then we r scaling it either through the range which is (max-min) or standard deviation in DENOMINATOR for all 3 types of scaling discussed above.*

Scaling is always multiplication, Shiting is always additive or subtractive.

We do not do scaling for one-hot encoded columns cz scaling may distort the binary nature of the one-hot encoded variables and make them less informative to the model.

### d. Robust Scaling

Useful when presence of outliers in data.

**Note**

*In all types of ML operations, we always do fit_transform on training data and only do transform on test data inorder to avoid data leakage problem. For e.g. if mean is required to be calculated for mean scaling then mean is only taken from training data during fit and it is just used on testing data to transform.*

# <font color=blue>**2. Feature Encoding (for categorical data)**

### a. **One hot encoding** 
Some image datasets are also one-hot encoded.

### b. **Label encoding** 

Replacing catogories with numerical values. Not easy to interpret o/p of label encoding. e.g. Monday, Tues either can be 1 or 2.

### c. **Lossy encoding** 

Choosing a handful important columns and put rest as Others by applying domain expertise. This is step up from one-hot encoding.

If not domain expert then only option is apply one hot encoding to all columns & then apply PCA.

Generally PCA recommended after one-hot encoding but if large number of categorical variables can apply PCA before one hot encoding as well.

# <font color=blue>**3. Feature Binning**

Bins are your categories. You are converting your numeric values to categories. We do this cz sometimes you have huge dataset like 20,000 data points and e.g. age values can vary from 0 to 100 then u can just use 10 bins for this data. e.g sometimes need to market some products to targeted customers of certain age group only.

Above K is number of bins you will have based on N which is number of samples you have. e.g. if you have 100 train data points, you can create 10 bins based on Square-root rule.

*At the end holy grail is better generalization of problem so whichever tools help u achieve that can be used. Its not that u have to use binning, one-hot encoding, etc, choose techniques & tools suited to your problem.The idea of this exercise is u should be aware u can engineer ur features in different ways*

### *<b>Note on Hyperparameter Tuning</b>*

<font color=blue>Model training with different hyperparameters is still done on training set but evaluation & best hyperparameters picked based on performance of model on the Validation set. Hence we say that hyperparameter tuning is done based on validation set.

In y=mx+c, m is a parameter that algorithm needs to determine based on training but things like learning rate, epochs are hyperparameters, you initially set a value & then tune these values so algorithm can find optimal values of parameters like m and c. Hyperparameters helps to control how model is learning.
